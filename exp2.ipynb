a)matrix operation (using vectorization)


```python
import numpy as np
m1=np.array([[1,3],[2,6]])
m2=np.array([[0,1],[1,9]])
print("Original 2-d matrix:")
print(m1)
print(m2)

print("Outer product of the two matrix is: ")
result=np.outer(m1,m2)
print(result)
```

b)Transformation using python

```python
m=[[1,2],[3,4],[5,6]]
for row in m:
  print(row)
res=[[m[j][i] for j in range(len(m))] for i in range(len(m[0]))]
print("\n")
for row in res:
  print(row)
```
